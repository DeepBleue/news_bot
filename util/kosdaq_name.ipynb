```python
from pykrx import stock

kosdaq_dict = {}

for ticker in stock.get_market_ticker_list(market='KOSDAQ'):
    name = stock.get_market_ticker_name(ticker)

    kosdaq_dict[name] = ticker

# print(kosdaq_dict)
```

```python
print("{")
for name, ticker in kosdaq_dict.items():
    print(f"    '{name}': '{ticker}',")
print("}")
```
